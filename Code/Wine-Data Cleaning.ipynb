```python
#@title Replace VOLATILE ACIDITY, CITRIC ACID, RESIDUAL SUGAR, CHLORIDES, FREE SULFUR DIOXIDE, DENSITY, PH, SULPHATES outliers with rules: <=0 : Q1 – 1.5 x IQR AND >= Q3 +1.5 x IQR: Q3 +1.5*IQR
```

```python
#name only columns that have outliers
outlier_column = df.columns.drop(["category","fixed acidity","alcohol","quality"])
```

```python
outlier_column
```

```python
upper_list = {}
lower_list = {}
```

```python
for i in outlier_column:
  Q1 = np.percentile(df[i], 25)
  Q3 = np.percentile(df[i], 75)  
  IQR = Q3 - Q1

  upper=Q3+1.5*IQR
  lower=Q1-1.5*IQR

  if lower <=0:
    lower = min(df[df[i]>=0][i])

  upper_list[i]=upper
  lower_list[i]=lower
```

```python
upper_list
```

```python
df['volatile acidity'] = np.where(df['volatile acidity'] > upper_list['volatile acidity'], upper_list['volatile acidity'], df['volatile acidity'])
df['volatile acidity'] = np.where(df['volatile acidity'] < lower_list['volatile acidity'], lower_list['volatile acidity'], df['volatile acidity'])

df['citric acid'] = np.where(df['citric acid'] > upper_list['citric acid'], upper_list['citric acid'], df['citric acid'])
df['citric acid'] = np.where(df['citric acid'] < lower_list['citric acid'], lower_list['citric acid'], df['citric acid'])

df['residual sugar'] = np.where(df['residual sugar'] > upper_list['residual sugar'], upper_list['residual sugar'], df['residual sugar'])
df['residual sugar'] = np.where(df['residual sugar'] < lower_list['residual sugar'], lower_list['residual sugar'], df['residual sugar'])

df['chlorides'] = np.where(df['chlorides'] > upper_list['chlorides'], upper_list['chlorides'], df['chlorides'])
df['chlorides'] = np.where(df['chlorides'] < lower_list['chlorides'], lower_list['chlorides'], df['chlorides'])

df['free sulfur dioxide'] = np.where(df['free sulfur dioxide'] > upper_list['free sulfur dioxide'], upper_list['free sulfur dioxide'], df['free sulfur dioxide'])
df['free sulfur dioxide'] = np.where(df['free sulfur dioxide'] < lower_list['free sulfur dioxide'], lower_list['free sulfur dioxide'], df['free sulfur dioxide'])

df['total sulfur dioxide'] = np.where(df['total sulfur dioxide'] > upper_list['total sulfur dioxide'], upper_list['total sulfur dioxide'], df['total sulfur dioxide'])
df['total sulfur dioxide'] = np.where(df['total sulfur dioxide'] < lower_list['total sulfur dioxide'], lower_list['total sulfur dioxide'], df['total sulfur dioxide'])

df['density'] = np.where(df['density'] > upper_list['density'], upper_list['density'], df['density'])
df['density'] = np.where(df['density'] < lower_list['density'], lower_list['density'], df['density'])

df['pH'] = np.where(df['pH'] > upper_list['pH'], upper_list['pH'], df['pH'])
df['pH'] = np.where(df['pH'] < lower_list['pH'], lower_list['pH'], df['pH'])

df['sulphates'] = np.where(df['sulphates'] > upper_list['sulphates'], upper_list['sulphates'], df['sulphates'])
df['sulphates'] = np.where(df['sulphates'] < lower_list['sulphates'], lower_list['sulphates'], df['sulphates'])
```

```python
#@title Handle Missing Values and Replace Outliers of Target Feature (Quality) with rules: <= 5 is low, ==6 is average, and >=7 is high
```

```python
df=df.dropna(axis=0)
```

```python
df.groupby('quality').size().plot(kind='bar')
```

```python
low = np.unique(df[df['quality'] <= 5]['quality'].values)
average = np.unique(df[df['quality'] == 6]['quality'].values)
high = np.unique(df[df['quality'] >=7]['quality'].values)
```

```python
#new_df = df.replace({'quality': {3:0,4:0,5:1,6:1,7:1,8:2,9:2}})
```

```python
df["quality"]=df["quality"].replace(low,0)
df["quality"]=df["quality"].replace(average,1)
df["quality"]=df["quality"].replace(high,2)
```

```python
sb.countplot(df,x='quality', palette = 'mako');
plt.title('Target Feature New Distribution');
```

```python
quality_count = df.groupby('quality').size()
```
