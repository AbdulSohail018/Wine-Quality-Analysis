```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_wine = pd.read_csv(r'C:\Users\abdul\Desktop\DATA 230\Project\raw_data.csv')
df_wine.head(10)
```

```python
import seaborn as sbn
corr_matrix = df_wine.corr()
plt.figure(figsize=(12,7))
sbn.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
plt.figure(figsize = (10,15))
for i, col in enumerate(df_wine.columns[1:].values):
    plt.subplot(4,3,i+1)
    df_wine.boxplot(col)
    plt.grid()
    plt.tight_layout()
```

```python
plt.figure(figsize = (10,15))
for i, col in enumerate(df_wine.columns[1:].values):
    plt.subplot(4,3,i+1)
    sbn.distplot(df_wine[col],kde = True, label = 'data')
    plt.grid()
    plt.legend()
    plt.tight_layout()
```
