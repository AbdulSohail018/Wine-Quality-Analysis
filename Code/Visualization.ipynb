```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

# 1. Correlogram of cleaned data

```python
corr_matrix = df.corr()
plt.figure(figsize=(12,7))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

# 2. Countplot of Winetype

```python
sns.countplot(x="quality", hue="category", data=df, palette="rocket")
plt.title('Countplot of Winetype')
```

# 3. Side by Side Bar Chart comparing Mean Components by WineType

```python
grouped = df.groupby('category')[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']].mean()
n = len(grouped.columns)
ind = np.arange(n)
width = 0.35
fig, ax = plt.subplots(figsize=(25,10))
rects1 = ax.bar(ind - width/2, grouped.loc['red'], width, label='Red Wine')
rects2 = ax.bar(ind + width/2, grouped.loc['white'], width, label='White Wine')
ax.set_xticks(ind)
ax.set_xticklabels(grouped.columns)
ax.set_xlabel('Wine Component')
ax.set_ylabel('Mean Value')
ax.set_title('Mean Values of Wine Components by Wine Type')
ax.legend()
plt.show()
```

# 4. Violin Plot for Wine Quality

```python
sns.violinplot(x='quality', y='alcohol', data=df)
plt.title('Violin Plot of Alcohol Content by Wine Quality')
plt.show()
```

# 5. Facetgrid for Wine Desnity vs quality by Wine Type

```python
g = sns.FacetGrid(df, col='category')
g = g.map(sns.kdeplot, 'quality', color = 'darkblue')
plt.subplots_adjust(top= 0.8)
g.fig.suptitle('Wine quality based on type of wine')
```

# 6. Scatter plot for Alcohol vs. Density

```python
sns.scatterplot(x="alcohol", y="density", data=df)
plt.title('Scatter Plot of Alcohol Content by Wine Density')
plt.show()
```

# 7. Swarm plot for Volatile Acidity by Wine Quality

```python
sns.swarmplot(x='quality', y='volatile acidity', data=df)
plt.title('Swarm Plot of Volatile Acidity by Wine Quality')
plt.show()
```
